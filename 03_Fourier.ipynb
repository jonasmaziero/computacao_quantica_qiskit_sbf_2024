# Transformada de Fourier quântica

A transformada de Fourier quântica é uma subrotina essencial para os principais algoritmos da computação quântica.

A transformada de Fourier discreta clássica é um mapa entre o conjunto de $d$ números complexos
$$x=(x_{0},\cdots,x_{d-1})$$
e o conjunto de $d$ números complexos
$$y=(y_{0},\cdots,y_{d-1})$$
definidos por
\begin{equation}
y_{k} := \frac{1}{\sqrt{d}}\sum_{j=0}^{d-1}x_{j}e^{2\pi ijk/d}.
\end{equation}

No caso quântico, esses conjuntos de números complexos são representados por estados quânticos. Mais precisamente, eles são os coeficientes de um vetor de estado quando representado na base computacional (de um qudit):
\begin{align}
|x\rangle & = \sum_{j=0}^{d-1}x_{j}|j\rangle = \begin{bmatrix} x_{0} \\ \vdots \\ x_{d-1} \end{bmatrix}, \\
|y\rangle & = \sum_{k=0}^{d-1}y_{k}|k\rangle = \begin{bmatrix} y_{0} \\ \vdots \\ y_{d-1} \end{bmatrix} \\
& \equiv \sum_{k=0}^{d-1}\Big(\frac{1}{\sqrt{d}}\sum_{j=0}^{d-1}x_{j}e^{2\pi ijk/d}\Big)|k\rangle \\
& = \sum_{j=0}^{d-1}x_{j}\Big(\frac{1}{\sqrt{d}}\sum_{k=0}^{d-1}e^{2\pi ijk/d}|k\rangle\Big).
\end{align}

Acima e doravamente, a menos que dito de outra forma, as representações matriciais de vetores de de operadores são feitas na base computacional
$$\{|0\rangle,|1\rangle,\cdots,|d-2\rangle,|d-1\rangle\}.$$

Vale observar que a codificação em estados quânticos dos números envolvidos na transformada de Fourier implicam nas seguintes restrições sobre esses números:
\begin{align}
& \sum_{j=0}^{d-1}|x_{j}|^{2} = |x_{0}|^{2}+|x_{1}|^{2}+\cdots+|x_{d-1}|^{2}=1, \\
& \sum_{j=0}^{d-1}|y_{j}|^{2} = |y_{0}|^{2}+|y_{1}|^{2}+\cdots+|y_{d-1}|^{2}=1,
\end{align}
e $$0\le|x_j|\le 1,\ 0\le|y_k|\le 1.$$
Vale observar também que se essas propriedades não forem satisfeitas, podemos renormalizar os números e extender o conjunto para que satisfaçam essas restrições.

Vemos que a transformação de estados mostrada acima,
\begin{align}
|y\rangle & =: U_{tfq}|x\rangle \\
& = U_{tfq}\sum_{j=0}^{d-1}x_{j}|j\rangle \\
& = \sum_{j=0}^{d-1}x_{j}U_{tfq}|j\rangle,
\end{align}
pode ser obtida se a ação da transformada de Fourier quântica na base computacional for
$$U_{tfq}|j\rangle = \frac{1}{\sqrt{d}}\sum_{k=0}^{d-1}e^{2\pi ijk/d}|k\rangle =: |F_j\rangle$$
para $j=0,\cdots,d-1$. Isso pode ser obtido com
$$U_{tfq} = \frac{1}{\sqrt{d}}\sum_{k,l=0}^{d-1}e^{2\pi ikl/d}|k\rangle\langle l| =: F.$$
$F$ é também chamada de porta de Fourier. Os estados $|F_j\rangle$ são chamados de estados de Fourier.

Vamos representar os vetores da base computacional de um qudit em termos da base computacional produto de
$$n=\log_{2}(d)$$
qubits. Ou seja, assumiremos que
$$d=2^{n}.$$

Ou seja, escreveremos
\begin{align}
|j\rangle & = |j_{1}j_2\cdots j_{n-1}j_{n}\rangle \\
& \equiv |j_{1}\rangle\otimes|j_2\rangle\otimes\cdots\otimes|j_{n-1}\rangle\otimes|j_{n}\rangle
\end{align}
com
$$j_{s}=0,1$$
para $s=1,\cdots,n$ e a representação decimal é obtida da representação binária como segue
\begin{align}
j & = j_{1}2^{n-1}+j_{2}2^{n-2}+\cdots+j_{n-1}2^{1}+j_{n}2^{0} \\
& = \sum_{l=1}^{n}j_{l}2^{n-l} \\
& = 0,1,\cdots,2^n-1.
\end{align}
Para uso posterior, notemos que
$$\frac{j}{2^{n}} = j2^{-n} = \sum_{l=1}^{n}j_{l}2^{-l}.$$

Seguindo, vamos atuar o operador de Fourier de um qudit em um estado da base computacional representado por $\log_{2}(d)$ qubits:
\begin{align}
& |F_j\rangle = F|j\rangle = F|j_{1}\cdots j_{n}\rangle \\
& = F(|j_{1}\rangle\otimes|j_{2}\rangle\otimes\cdots\otimes|j_{n-1}\rangle\otimes|j_{n}\rangle) \\
& = \frac{1}{\sqrt{d}}\sum_{k=0}^{d-1}e^{2\pi ijk/d}|k\rangle \\
& = \frac{1}{\sqrt{2^{n}}}\sum_{k=0}^{2^{n}-1}e^{2\pi ijk/2^{n}}|k\rangle \\
& = \frac{1}{\sqrt{2^{n}}}\sum_{k_{1},k_{2},\cdots, k_{n-1},k_{n}=0}^{1}e^{2\pi ij\sum_{l=1}^{n}k_{l}2^{-l}}|k_{1}k_{2}\cdots k_{n-1}k_{n}\rangle  \\
& = \frac{1}{\sqrt{2^{n}}}\sum_{k_{1},k_{2},\cdots,k_{n-1},k_{n}=0}^{1}\Pi_{l=1}^{n}e^{2\pi ijk_{l}2^{-l}}|k_{1}k_{2}\cdots k_{n-1}k_{n}\rangle \\
& = \frac{1}{\sqrt{2^{n}}}\sum_{k_{1},k_{2},\cdots,k_{n-1},k_{n}=0}^{1}e^{2\pi ijk_{1}2^{-1}}e^{2\pi ijk_{2}2^{-2}}\cdots e^{2\pi ijk_{n-1}2^{-n+1}}e^{2\pi ijk_{n}2^{-n}}|k_{1}\rangle\otimes|k_{2}\rangle\otimes\cdots\otimes|k_{n-1}\rangle\otimes|k_{n}\rangle \\
& = \frac{1}{\sqrt{2^{n}}}\Big(\sum_{k_{1}=0}^{1}e^{2\pi ijk_{1}2^{-1}}|k_{1}\rangle\Big)\otimes\Big(\sum_{k_{2}=0}^{1}e^{2\pi ijk_{2}2^{-2}}|k_{2}\rangle\Big)\otimes\cdots \otimes\Big(\sum_{k_{n-1}=0}^{1}e^{2\pi ijk_{n-1}2^{-n+1}}|k_{n-1}\rangle\Big)\otimes\Big(\sum_{k_{n}=0}^{1}e^{2\pi ijk_{n}2^{-n}}|k_{n}\rangle\Big) \\
& = \frac{1}{\sqrt{2^{n}}}\Big(|0\rangle+e^{2\pi ij2^{-1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-2}}|1\rangle\Big)\otimes\cdots \otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big). \hspace{1cm} (\star)
\end{align}

Ou seja, para implementar a TFQ, precisamos preparar $n$ qubits em um estado produto de estados de um qubit que são uma superposição dos estados da base computacional de um qubit com o estado $|1\rangle$ do qubit $p$ recebendo uma fase
$$e^{2\pi ij2^{-p}}$$
com
$$p=1,2,\cdots,n.$$
Podemos escrever essa fase como segue
\begin{align}
e^{2\pi ij2^{-p}} & = e^{2\pi i\sum_{l=1}^{n}j_{l}2^{n-l}2^{-p}} \\
& = e^{2\pi i\sum_{l=1}^{n}j_{l}2^{n-l-p}} \\
& = \Pi_{l=1}^{n}e^{2\pi ij_{l}2^{n-l-p}} \\
& = e^{2\pi ij_{1}2^{n-1-p}}e^{2\pi ij_{2}2^{n-2-p}}\cdots e^{2\pi ij_{n-1}2^{n-n+1-p}}e^{2\pi ij_{l}2^{n-n-p}} \\
& = e^{2\pi ij_{1}2^{n-1-p}}e^{2\pi ij_{2}2^{n-2-p}}\cdots e^{2\pi ij_{n-1}2^{1-p}}e^{2\pi ij_{n}2^{-p}}.
\end{align}
Alguns casos particulares são:
\begin{align}
& e^{2\pi ij2^{-1}} = e^{2\pi ij_{1}2^{n-2}}e^{2\pi ij_{2}2^{n-3}}\cdots e^{2\pi ij_{n-1}2^{0}}e^{2\pi ij_{n}2^{-1}}, \\
& e^{2\pi ij2^{-2}} = e^{2\pi ij_{1}2^{n-3}}e^{2\pi ij_{2}2^{n-4}}\cdots e^{2\pi ij_{n-1}2^{-1}}e^{2\pi ij_{n}2^{-2}}, \\
& \vdots \\
& e^{2\pi ij2^{-n+1}} = e^{2\pi ij_{1}2^{0}}e^{2\pi ij_{2}2^{-1}}\cdots e^{2\pi ij_{n-1}2^{2-n}}e^{2\pi ij_{n}2^{-n+1}}, \\
& e^{2\pi ij2^{-n}} = e^{2\pi ij_{1}2^{-1}}e^{2\pi ij_{2}2^{-2}}\cdots e^{2\pi ij_{n-1}2^{1-n}}e^{2\pi ij_{n}2^{-n}}.
\end{align}

A ação da porta de Hadamard pode ser escrita de forma geral como
$$H|j_{l}\rangle = \frac{1}{\sqrt{2}}\big(|0\rangle+e^{2\pi ij_{l} 2^{-1}}|1\rangle\big),$$
pois
\begin{align}
& j_l=0\ \therefore\ e^{2\pi i(0) 2^{-1}}=e^{0} = 1, \\
& j_l=1\ \therefore\ e^{2\pi i(1) 2^{-1}}=e^{\pi i} = -1.
\end{align}

A porta de fase,
$$P(\phi) = |0\rangle\langle 0|+e^{i\phi}|1\rangle\langle 1|,$$
atua como segue:
$$P(\phi)|0\rangle = |0\rangle \text{ e } P(\phi)|1\rangle = e^{i\phi}|1\rangle.$$
Olhando para a última igualdade para o estado $|F_j\rangle$, vemos que a forma dos estados locais se parecem com aquelas que podemos obter aplicando uma porta de fase $P(\phi)$. No entanto, vemos também que essas fases, $e^{2\pi ij2^{-p}}$, dependem dos estados iniciais de todos os qubits. Isso nos indica que precisaremos aplicar portas de fases controladas
$$C_{P(\phi)}^{A\rightarrow B} = |0\rangle_{A}\langle 0|\otimes\mathbb{I}_{B} + |1\rangle_{A}\langle 1|\otimes P(\phi).$$
Ou seja, a porta de fase dá uma fase $e^{i\phi}$ ao target se o estado do controle for $|1\rangle$. Ou poderíamos dizer que essa porta adiciona uma fase
$$e^{i\phi j_{p}}$$
ao estado $|1\rangle$ target, com $j_{p}=0,1$ dependendo do estado do qubit target $p$.

Vamos começar olhando para o estado do 1º qubit, que é proporcional a
$$|0\rangle+e^{2\pi ij2^{-1}}|1\rangle.$$
Essa exponencial é dada por
\begin{align}
e^{2\pi ij2^{-1}} & = e^{2\pi ij_{1}2^{n-2}}e^{2\pi ij_{2}2^{n-3}}\cdots e^{2\pi ij_{n-1}2^{0}}e^{2\pi ij_{n}2^{-1}} \\
& = e^{2\pi i\big(j_{1}2^{n-2}+j_{2}2^{n-3}\cdots+j_{n-1}2^{0}+j_{n}2^{-1}\big)}.
\end{align}
Pelo $j_{1}$, vamos começar aplicando a Hadamard no 1º qubit
$$(H_{1}|j_{1}\rangle)\otimes|j_{2}\cdots j_{n}\rangle = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij_{1}2^{-1}}|1\rangle\Big)\otimes|j_{2}\cdots j_{n}\rangle$$

Seguindo, aplicaremos a porta de fase controlada com o controle indo, individualmente, desde o qubit $l=2$ até o qubit $l=n$, adicionando uma fase
$$e^{2\pi ij_{l}2^{-l}}$$
ao estado $|1\rangle$ do target, o qubit 1 neste caso.

Para o qubit $2$ como controle teremos
\begin{align}
& C_{P(2\pi i2^{-2})}^{2\rightarrow 1}(H_{1}|j_{1}\rangle)\otimes|j_{2}\rangle|j_3\cdots j_{n}\rangle \\
& = \big(I\otimes|0\rangle_2\langle 0| + P(2\pi i2^{-2})\otimes|1\rangle_2\langle 1|\big)\frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi i j_{1}2^{-1}}|1\rangle\Big)\otimes|j_{2}\rangle|j_3\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi i(j_{1}2^{-1}+j_{2}2^{-2})}|1\rangle\Big)\otimes|j_{2}\rangle|j_3\cdots j_{n}\rangle.
\end{align}
Note que se $j_2=0$ o estado não muda. Se $j_2=1$ o estado $|0\rangle$ não muda e o estado $|1\rangle$ recebe uma fase $e^{2\pi ij_{2}2^{-2}}.$

Como a ação para os outros qubits como controle é similar, teremos
\begin{align}
& \Pi_{q=2}^{n}C_{P(2\pi i2^{-q})}^{q\rightarrow 1}(H_{1}|j_{1}\rangle)\otimes|j_{2}j_3\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi i(j_{1}2^{-1}+j_{2}2^{-2}+\cdots+j_{n}2^{-n})}|1\rangle\Big)\otimes|j_{2}j_3\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi i(j_{1}2^{n-1}+j_{2}2^{n-2}+\cdots+j_{n}2^{n-n})2^{-n}}|1\rangle\Big)\otimes|j_{2}j_3\cdots j_{n}\rangle \\
&  = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes|j_{2}j_3\cdots j_{n}\rangle \\
& =: |\xi_{0}\rangle.
\end{align}
Note que este é o estado do qubit $n$ no estado $F|j\rangle$. Isso indica que depois de finalizar as operações, precisaremos fazer uma troca entre os estados dos qubits $1$ e $n$.

Continuando, aplicamos uma Hadamard no 2º qubit, obtendo
$$H_{2}|\xi_{0}\rangle = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij_{2}2^{-1}}|1\rangle\Big)\otimes|j_{3}\cdots j_{n}\rangle.$$

Na sequência, aplica-se portas de fase controladas como segue
\begin{align}
& \Pi_{q=3}^{n}C_{P(2\pi i2^{-q+1})}^{q\rightarrow 2}\frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\big(H_{2}|j_{2}\rangle\big)\otimes|j_{3}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi i(j_{2}2^{-1}+j_{3}2^{-2}+\cdots+j_{n}2^{-n+1})}|1\rangle\Big)\otimes|j_{3}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi i(j_{1}2^{0}+j_{2}2^{-1}+j_{3}2^{-2}+\cdots+j_{n}2^{-n+1})}|1\rangle\Big)\otimes|j_{3}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi i(j_{1}2^{n-1}+j_{2}2^{n-2}+j_{3}2^{n-3}+\cdots+j_{n}2^0)2^{-n+1}}|1\rangle\Big)\otimes|j_{3}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes|j_{3}\cdots j_{n}\rangle \\
& =: |\xi_{1}\rangle,
\end{align}
em que usamos e
\begin{align}
e^{2\pi ij_{1}2^{0}} & = e^{2\pi ij_{1}} = \cos(j_{1}2\pi)+i\sin(j_{1}2\pi) = 1+i0 \\
& = 1
\end{align}
para $j_{1}=0,1$.

Fazendo mais um passo, aplicamos a Hadarmad no 3º qubit e as portas de fase controladas no qubit 3 com os qubits controle indo desde o 4º até o n-ésimo. Com isso, obteremos
\begin{align}
& \Pi_{q=4}^{n}C_{P(2\pi i2^{-q+2})}^{q\rightarrow 3}\frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(H_{3}|j_{3}\rangle\Big)\otimes|j_{4}j_{5}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi i(j_{3}2^{-1}+j_{4}2^{-2}+\cdots+j_{n}2^{-n+2})}|1\rangle\Big)\otimes|j_{4}j_{5}\cdots j_{n}\rangle \\
& \equiv \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi i(j_{1}2^{1}+j_{2}2^{0}+j_{3}2^{-1}+j_{4}2^{-2}+\cdots+j_{n}2^{-n+2})}|1\rangle\Big)\otimes|j_{4}j_{5}\cdots j_{n}\rangle \\
& = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+2}}|1\rangle\Big)\otimes|j_{4}j_{5}\cdots j_{n}\rangle,
\end{align}
em que usamos
\begin{align}
& e^{2\pi ij_{2}2^{0}} = e^{2\pi ij_{2}} = \cos(j_{2}2\pi)+i\sin(j_{2}2\pi) = 1, \\
& e^{2\pi ij_{1}2^{1}} = e^{4\pi ij_{1}} = \cos(j_{1}4\pi)+i\sin(j_{1}4\pi) = 1
\end{align}
para $j_{1},j_{2}=0,1$
\begin{align}
j2^{-n+2} & = \sum_{l=1}^{n}j_{l}2^{n-l}2^{-n+2} = \sum_{l=1}^{n}j_{l}2^{-l+2} \\
& = j_{1}2^{1} + j_{2}2^{0} + j_{3}2^{-1} + \cdots + j_{n-1}2^{-n+3} + j_{n}2^{-n+2}.
\end{align}

Bom, com isso podemos inferir que se continuarmos com esse processo, de aplicar a porta de Hadamard $H_{l}$ no qubit $l=4,\cdots,n$ seguida de uma fase controlada aplicada no qubit $l$ com o qubit $q=l+1,\cdots,n$ como controle,
$$\Pi_{q=l+1}^{n}C_{P(2\pi i2^{-q+l-1})}^{q\rightarrow l},$$
obteremos o seguinte estado
$$ |\Xi\rangle = \frac{1}{\sqrt{2}}\Big(|0\rangle+e^{2\pi ij2^{-n}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+1}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-n+2}}|1\rangle\Big)\otimes\cdots\otimes\Big(|0\rangle+e^{2\pi ij2^{-2}}|1\rangle\Big)\otimes\Big(|0\rangle+e^{2\pi ij2^{-1}}|1\rangle\Big).$$
Observe que para o qubit $n$ só aplicamos a Hadamard, sem aplicação de fases controladas.

Vemos que este estado é equivalente ao estado da eq. ($\star$) se trocarmos os estados dos qubit 1 e $n$, $2$ e $n-1$, $3$ e $n-2$, $\cdots$. Assim, finalmente obtemos que
\begin{equation}
F|j\rangle = \Pi_{l=1}^{n/2}SWAP^{l\leftrightarrow n-l+1}|\Xi\rangle.
\end{equation}
Essa equação é válida se $n$ for par. Se $n$ for ímpar, devemos usar como limite superior do produtório:
$$(n-1)/2.$$
Lembrando
\begin{align}
SWAP^{l\leftrightarrow n-l+1} & = CNOT^{l\rightarrow n-l+1}CNOT^{n-l+1\rightarrow l}CNOT^{l\rightarrow n-l+1} \\
& = CNOT^{n-l+1\rightarrow l}CNOT^{l\rightarrow n-l+1}CNOT^{n-l+1\rightarrow l}.
\end{align}

Vale observar que essas SWAPs comutam e, a princípio, poderiam ser aplicadas em paralelo. Mas, na prática isso não é bem assim pois não temos conectividade entre todos os qubits do chip, e isso implica em várias SWAPs intermedidárias de estado para então aplicar as CNOTs entre os qubits $l$ e $n-l+1.$

```python
import numpy as np
from qiskit import QuantumCircuit
```

O circuito quântico para a transformada de Fourier quântica está implementado e exemplificado com o código abaixo.

```python
def qc_qft(n):
    qc = QuantumCircuit(n, name='QFT')
    for l in range(0, n):
        qc.h(l)
        if l < n-1:
            for q in range(l+1, n):
                lb = 2*np.pi*2**(-q+l-1)
                qc.cp(lb, q, l)
            #qc.barrier()
    #qc.barrier()
    if n%2 == 0:
        ul = n//2
    else:
        ul = (n-1)//2
    for p in range(0, ul):
        qc.swap(p, n-1-p)
    return qc
```

```python
qcqft = qc_qft(4)
qcqft.draw('mpl')
```

# TFQ inversa
Para obter o circuito quântico para a transformada de Fourier quântica inversa, notemos primeiramente que a inversa da porta de Hadamard é a própria porta de Hadamard:
$$H^{\dagger} = H.$$
Como a CNOT também é hermitiana, teremos também que
$$SWAP^{\dagger} = SWAP$$
pois
$$SWAP_{AB} = CNOT_{AB}CNOT_{BA}CNOT_{AB}$$
e a $CNOT$ é hermitiana, i.e., $CNOT=CNOT^\dagger$ .

Já para a porta de fase temos que
\begin{align}
P(\phi)^{\dagger} & = \big(|0\rangle\langle 0| + e^{i\phi}|1\rangle\langle 1|\big)^\dagger \\
& = |0\rangle\langle 0| + e^{-i\phi}|1\rangle\langle 1| \\
& = P(-\phi).
\end{align}
Assim
\begin{align}
C_{P(\phi)}^{\dagger} & = \big(|0\rangle\langle 0|\otimes I + |1\rangle\langle 1|\otimes P(\phi)\big)^\dagger \\
 & = |0\rangle\langle 0|\otimes I + |1\rangle\langle 1|\otimes P(-\phi) \\
& = C_{P(-\phi)}.
\end{align}
Assim, como implementado e exemplificado com o código abaixo, obtemos $F^\dagger$ aplicando as operações inversas que aparecem na TFQ, da direita para a esquerda do circuito, de forma que
$$F^\dagger F=I.$$

```python
def qc_iqft(n):
    qc = QuantumCircuit(n, name='IQFT')
    if n%2 == 0:
        ul = n//2
    else:
        ul = (n-1)//2
    for p in range(ul-1, -1, -1):
        qc.swap(p, n-1-p)
    #qc.barrier()
    for l in range(n-1, -1, -1):
        if l < n-1:
            for q in range(n-1, l+1-1, -1):
                lb = -2*np.pi*2**(-q+l-1)
                qc.cp(lb, q, l)
        qc.h(l)
        #qc.barrier()
    return qc
```

```python
qciqft = qc_iqft(4)
qciqft.draw('mpl')
```

## Testes básicos

### Preparação do estado $|F_1\rangle = F|1\rangle$

Como um teste básico, vamos considerar $n=3\ \therefore\ d=8$ e o estado inicial $|001\rangle \equiv |1\rangle$. Nesse caso, temos que
\begin{align}
F|1\rangle & = \frac{1}{\sqrt{8}}\sum_{k=0}^7 e^{2\pi i (1)k/d}|k\rangle \\
& = \frac{1}{\sqrt{8}}\Big(e^{2\pi i(0)/8}|0\rangle + e^{2\pi i(1)/8}|1\rangle + e^{2\pi i(2)/8}|2\rangle + e^{2\pi i(3)/8}|3\rangle + e^{2\pi i(4)/8}|4\rangle + e^{2\pi i(5)/8}|5\rangle + e^{2\pi i(6)/8}|6\rangle + e^{2\pi i(7)/8}|5\rangle\Big) \\
& = \frac{1}{\sqrt{8}}\Big(|0\rangle + e^{\pi i/4}|1\rangle + e^{\pi i/2}|2\rangle + e^{3\pi i/4}|3\rangle + e^{\pi i}|4\rangle + e^{5\pi i/4}|5\rangle + e^{3\pi i/2}|6\rangle + e^{7\pi i/4}|5\rangle\Big).
\end{align}

### Probabilidade simulada e experimental em função do número de qubits

```python
qc_list = []
qb_list = []
nqb_max = 22
for n in range(1, nqb_max, 4):
  qb_list.append(n)
  qc = QuantumCircuit(n,n)
  qc.x(n-1)
  #qc.barrier()
  qcqft = qc_qft(n)
  qc.append(qcqft, list(range(n)))
  #qc.barrier()
  qciqft = qc_iqft(n)
  qc.append(qciqft, list(range(n)))
  #qc.barrier()
  qc.measure(list(range(n)),list(range(n)))

  qc_list.append(qc.decompose())
```

```python
qc_list[0].draw('mpl')
```

#### Simulação

```python
from qiskit_aer import AerSimulator
backend = AerSimulator()
```

```python
from qiskit_ibm_runtime import SamplerV2 as Sampler
sampler = Sampler(backend=backend)
```

```python
nshots = 2**12 # = 4096
job = sampler.run(qc_list, shots=nshots)
```

```python
counts_sim_list = []
prob_1_sim_list = []
for j in range(0,len(qc_list)):
  counts_sim_list.append(job.result()[j].data.c.get_counts())
  for k in range(1, nqb_max, 4):
    st = '1' + '0'*(k-1)
    if st in counts_sim_list[j]:
      prob_1_sim_list.append(counts_sim_list[j][st]/nshots)
```
